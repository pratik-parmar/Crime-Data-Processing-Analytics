```python
# fixing date from type object to date time
# fixing Updates date on from type object to date time
import pandas as pd
all_chunks = []   # store processed chunks

for chunk in pd.read_csv(file_path, chunksize=1000):

    chunk["Updated On"] = pd.to_datetime(
    chunk["Updated On"],
    format="%m/%d/%Y %I:%M:%S %p",
    errors="coerce"
)



    chunk = chunk.dropna()

    chunk["Latitude"] = chunk["Latitude"].astype("float32")
    chunk["Longitude"] = chunk["Longitude"].astype("float32")
    chunk["Beat"] = chunk["Beat"].astype("int32")

    # 4. Store processed chunk
    all_chunks.append(chunk)


# 🔥 Combine all chunks
df = pd.concat(all_chunks)

# 🔥 Global sort by ID
df = df.sort_values(by="ID").reset_index(drop=True)

# View result
print(df.head())
print(len(df))


```
